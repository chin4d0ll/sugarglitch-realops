# ติดตั้งและใช้งาน Extensions ผ่าน Visual Studio Code
ติดตั้ง Extensions เหล่านี้เพื่อประสบการณ์การพัฒนาที่ดีที่สุด:
1. เปิด Extensions view (Ctrl+Shift+X)
2. ค้นหาชื่อ extension
3. คลิก Install

# แนะนำ Extensions สำหรับ Python Development
- **ms-python.python** - Python extension pack
- **ms-python.flake8** - Python linting
- **ms-python.black-formatter** - Code formatting
ตัวอย่างการตั้งค่า:
```json
{
  "python.linting.enabled": true,
  "python.formatting.provider": "black"
}
```

# แนะนำ Extensions สำหรับ Remote Development
- **ms-vscode-remote.remote-containers** - Dev Containers
- **ms-vscode-remote.remote-ssh** - Remote SSH
ตัวอย่างการเชื่อมต่อ:
```bash
# Remote SSH
ssh user@remote-server
```

# แนะนำ Extensions สำหรับ Database
- **cweijan.vscode-mysql-client2** - MySQL/PostgreSQL client
- **mtxr.sqltools** - Universal SQL tools
- **qwtel.sqlite-viewer** - SQLite viewer
ตัวอย่างการเชื่อมต่อฐานข้อมูล:
```sql
SELECT * FROM users;
```

# แนะนำ Extensions สำหรับ Utilities
- **eamodio.gitlens** - Git integration
- **humao.rest-client** - API testing
- **ms-toolsai.jupyter** - Jupyter notebooks
- **ms-vscode.vscode-json** - JSON support
- **github.copilot** - AI assistant
ตัวอย่างการใช้งาน:
```json
{
  "gitlens.advanced": true
}
```

# ตัวอย่างการติดตั้ง Extension ด้วยคำสั่งใน Terminal
ติดตั้ง Extensions ด้วยคำสั่งใน Terminal:
```bash
code --install-extension ms-python.python
code --install-extension ms-python.flake8
code --install-extension ms-python.black-formatter
```
